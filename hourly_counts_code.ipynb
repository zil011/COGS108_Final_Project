```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as skl
import datetime
import csv
```

```python
ubermay = pd.read_csv("uber-raw-data-may14.csv")
weatherdesc = pd.read_csv("weather_description_clean.csv")
```

```python
ubermay2 = ubermay.head(250)
print(ubermay2)
```

```python
count = 0
if "5/{}/2014 {}:".format(1,0) in ubermay2.iloc[0,0]:
    count += 1
print(count)
```

```python
counts = {}
for date in np.arange(1,32):
    for time in np.arange(0,24):
        counts["5/{}/2014 {}".format(date,time)] = 0
```

```python
for index, row in ubermay.iterrows():
        counts[row[0][:row[0].index(":")]] += 1
```

```python
print(counts)
```

```python
with open('uber_hourly_counts_may.csv', 'w') as f:
    for key in counts.keys():
        f.write("%s,%s\n"%(key,counts[key]))
```
